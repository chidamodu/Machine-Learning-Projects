```python
def convert_radians(df, c1, c2):
    import math
    
    longitude_cos=[]
    longitude_sine=[]
    latitude_cos=[]
    latitude_sine=[]


    for i in df[c1]:
        longitude_cos.append(math.cos(round(math.radians(i), 4)))
        longitude_sine.append(math.sin(round(math.radians(i), 4)))
        
        
    for j in df[c2]:
        latitude_cos.append(math.cos(round(math.radians(j), 4)))
        latitude_sine.append(math.sin(round(math.radians(j), 4)))
        
       
         
    return longitude_cos, longitude_sine, latitude_cos, latitude_sine
```

```python
def formula_calculation(df, c1, c2, c3, c4):
    import numpy as np
    
    dlat1=[]
    clat1=[]
    cos_clat=[]
    cos_lat_rad=[]

    for i in df[c1]:#dlat
        dlat1.append((np.sin(i/2))**2)
        
    for j in df[c2]:#dlon
        clat1.append((np.sin(j/2))**2)
        
    for k in df[c3]:#central_latitude
        cos_clat.append(np.cos(k))
        
    for l in df[c4]:#latitude_radians
        cos_lat_rad.append(np.cos(l))
        
    return dlat1, clat1, cos_clat, cos_lat_rad
        
        
        
```

```python
#I have used radius=6371000 because the height is in meters and therefore converting every distance value to meters
#actually radius=6371 km

def haversine_distance(df, c, radius=6371000):
    import numpy as np
    distance=[]
    for a in df[c]:
        c=2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
        distance.append((radius*c))
        
    return pd.Series(distance)
        
    
    
```

```python
def get_postcode(df, c1, c2):
    from geopy.geocoders import Nominatim
    location=[]
    geolocator = Nominatim(user_agent="myGeocoder")
    location.append(([geolocator.reverse(x, y) for x, y in zip(df[c1], df[c2])]))
    return location
    
   
```
