```python
#getting the postcode of every latitude and longitude pair in order to bin the nearby postcodes and thereby use it as a 
#proxy to mapid

def finding_postcode(c1, c2):
    from geopy.geocoders import Nominatim
    postcode=[]
    geolocator = Nominatim(user_agent="myGeocoder")
    for i in list(zip(c1, c2)):
        location = geolocator.reverse(i)
        postcode.append(location.raw.get('address').get('postcode'))
        
    return postcode
     
```

```python
#Error!
#committed a mistake here. forgot to find math.cos() and math.sin() for longitude values

def convert_radians(df, c1, c2):
    import math
    
    longitude_cos=[]
    longitude_sine=[]
    latitude_cos=[]
    latitude_sine=[]


    for i in df[c1]:
        longitude_cos.append(round(math.radians(i), 4))
        longitude_sine.append(round(math.radians(i), 4))
        
        
    for j in df[c2]:
        latitude_cos.append(math.cos(round(math.radians(j), 4)))
        latitude_sine.append(math.sin(round(math.radians(j), 4)))
        
        

        
        
         
    return longitude_cos, longitude_sine, latitude_cos, latitude_sine
```
